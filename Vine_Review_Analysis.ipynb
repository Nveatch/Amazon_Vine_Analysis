```python
# Imports
import pandas as pd
```

```python
# Read in csv
vine_df= pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
# Filter for total_votes > 20
vine_total_votes_df = vine_df [vine_df.total_votes >= 20]
vine_total_votes_df.head()
```

```python
# Df where helpful_votes/total_votes >=0.5
vine_helpful_total_df = vine_total_votes_df[vine_total_votes_df.helpful_votes/vine_total_votes_df.total_votes >= 0.5]
vine_helpful_total_df.head()
```

```python
# df filter for paid reviews
vine_vine_df = vine_helpful_total_df[vine_helpful_total_df.vine =='Y']
vine_vine_df.head()
```

```python
# df filter for unpaid reviews
vine_no_vine_df = vine_helpful_total_df[vine_helpful_total_df.vine =='N']
vine_no_vine_df.head()
```

```python
# Create Analysis Dataframe
analysis_df = pd.DataFrame(columns=["Vine Type","Total Reviews","Number of 5-Star Reviews", "Percentage of 5-Star Reviews"])
```

```python
# Paid Analysis
paid_total_reviews=len(vine_vine_df)
paid_5_star =len(vine_vine_df[vine_vine_df.star_rating == 5])
paid_percentage = "{:.0%}".format(paid_5_star/paid_total_reviews)
analysis_df = analysis_df.append({"Vine Type":"Paid","Total Reviews":paid_total_reviews,"Number of 5-Star Reviews":paid_5_star, "Percentage of 5-Star Reviews":paid_percentage},ignore_index = True)
```

```python
# Unpaid Analysis
unpaid_total_reviews=len(vine_no_vine_df)
unpaid_5_star =len(vine_no_vine_df[vine_no_vine_df.star_rating == 5])
unpaid_percentage = "{:.0%}".format(unpaid_5_star/unpaid_total_reviews)
analysis_df = analysis_df.append({"Vine Type":"Unpaid","Total Reviews":unpaid_total_reviews,"Number of 5-Star Reviews":unpaid_5_star, "Percentage of 5-Star Reviews":unpaid_percentage},ignore_index = True)
```

```python
# Changing Index
analysis_df.set_index("Vine Type")
```
